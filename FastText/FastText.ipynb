```python
import pandas as pd

df= pd.read_csv("subset_tweets_preprocessed.csv")
df
```

```python
df['label'] = '__label__' + df['label'].astype(str)
df.head(-3)
```

```python
df['label_description'] = df['text'] + ' ' + df['label']
df.head(-3)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
```

```python
train.shape, test.shape
```

```python
train.to_csv("subset_tweets_preprocessed.train", columns=["label_description"], index=False, header=False)
test.to_csv("subset_tweets_preprocessed.test", columns=["label_description"], index=False, header=False)
```

```python
import fasttext

# Train the model
model = fasttext.train_supervised(input="subset_tweets_preprocessed.train")

# Evaluate the model
result = model.test("subset_tweets_preprocessed.test")

# Print the evaluation result
print(result)
```
